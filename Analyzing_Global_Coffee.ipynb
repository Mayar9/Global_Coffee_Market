# Task1

## task details

* Create a bar chart of the top 10 coffee producers by total production in 'total-production.csv', sorted from highest to lowest. Change the xticks to two letter country abbreviations.

#### '''Way1'''

```python
top10_producers = coffee_production.assign(overall_prod = coffee_production.iloc[:, 1:].sum(axis=1).round(2)).sort_values('overall_prod', ascending=False).iloc[:10, [0, -1]].set_index('total_production')
# top10_producers = coffee_production.assign(overall_prod = coffee_production.iloc[:, 1:].sum(axis=1).round(2)).groupby('total_production').agg({'overall_prod':'sum'}).sort_values('overall_prod', ascending=False)[:10]
top10_producers
# print(top10_producers.info())
# print(top10_producers.index)
labels = ['BR', 'VN', 'CO', 'ID', 'ET', 'IN','MX', 'GT', 'HN', 'UG']

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x=top10_producers.index, height=top10_producers['overall_prod']/1000000, width=0.6, color='skyblue')

# ax.set_xticks(top10_producers.index, top10_producers.index.str[:4].str.upper())
ax.set_xticks(top10_producers.index, labels)
ax.set_ylabel('Total Productions', fontsize=10)

ax.spines[['top', 'right']].set_visible(False)


fig.suptitle('Top Ten Coffee Producers', fontweight='bold', fontsize=14)
```

#### '''Way2'''

```python
coffee_T = coffee_production.T
coffee_T.columns = coffee_T.iloc[0, :]
coffee_T = coffee_T.drop('total_production', axis=0)
coffee_T.head()
```

```python
# coffee_T.sum()  #pandas series
top10_producers_V2 = coffee_T.sum(axis=0).sort_values(ascending=False)[:10]
top10_producers_V2

# print(top10_producers_V2.info())
labels = ['BR', 'VN', 'CO', 'ID', 'ET', 'IN','MX', 'GT', 'HN', 'UG']

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x=top10_producers_V2.index, height=top10_producers_V2/1000000, width=0.6, color='skyblue')

# ax.set_xticks(top10_producers_V2.index, top10_producers_V2.index.str[:4].str.upper())
ax.set_xticks(top10_producers_V2.index, labels)
ax.set_ylabel('Total Productions', fontsize=10)

ax.spines[['top', 'right']].set_visible(False)


fig.suptitle('Top Ten Coffee Producers', fontweight='bold', fontsize=14)
```

# Task2

## task details

* Create a line chart with the top 5 nations over time. Make sure to include a legend! Consider converting the index to datetime64[ns].

#### Way1 

```python
top5_producers_df = coffee_T[coffee_T.sum().sort_values(ascending=False)[:5].index]
top5_producers_df.index = top5_producers_df.index.astype('datetime64[ns]')
top5_producers_df

fig, ax = plt.subplots(figsize=(14,6))

ax.plot(top5_producers_df.index, top5_producers_df['Brazil'], )
ax.plot(top5_producers_df.index, top5_producers_df['Viet Nam'], )
ax.plot(top5_producers_df.index, top5_producers_df['Colombia'],)
ax.plot(top5_producers_df.index, top5_producers_df['Indonesia'], )
ax.plot(top5_producers_df.index, top5_producers_df['Ethiopia'],)


ax.set_title("Viet Nam Surges to Number 2 Spot")
ax.set_ylabel("Production (Thousand 60kg Bags)")
fig.suptitle("Top Coffee Producing Nations 1990-2018", fontweight='bold', fontsize=14)


ax.spines[['top', 'right']].set_visible(False)
# ax.xaxis.set_tick_params(rotation=45)


fig.legend(top5_producers_df.columns)
plt.show()
```

#### Way2 (Stacked Area chart)

```python
# top5_producers_df['Brazil'] = pd.to_numeric(top5_producers_df['Brazil'])
# top5_producers_df['Viet Nam'] = pd.to_numeric(top5_producers_df['Viet Nam'])
# top5_producers_df['Colombia'] = pd.to_numeric(top5_producers_df['Colombia'])
# top5_producers_df['Indonesia'] = pd.to_numeric(top5_producers_df['Indonesia'])
# top5_producers_df['Ethiopia'] = pd.to_numeric(top5_producers_df['Ethiopia'])

top5_producers_df.astype({"Brazil": "float64", "Viet Nam": "float64", "Colombia": "float64", "Indonesia": "float64", "Ethiopia": "float64"})

fig, ax = plt.subplots(figsize=(14,6))
# print(top5_producers_df.info())


ax.stackplot(
    top5_producers_df.index, 
    top5_producers_df['Brazil'],
    top5_producers_df['Viet Nam'], 
    top5_producers_df['Colombia'],
    top5_producers_df['Indonesia'],
    top5_producers_df['Ethiopia'],
    colors = ['blue', 'green', 'pink', 'orange', 'purple']
)


ax.set_title("Viet Nam Surges to Number 2 Spot")
ax.set_ylabel("Production (Thousand 60kg Bags)")
fig.suptitle("Top Coffee Producing Nations 1990-2018", fontweight='bold', fontsize=14)


ax.spines[['top', 'right']].set_visible(False)
# ax.xaxis.set_tick_params(rotation=45)


fig.legend(top5_producers_df.columns)

plt.show()
```
